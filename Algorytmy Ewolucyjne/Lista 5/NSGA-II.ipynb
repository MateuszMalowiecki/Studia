```python
import numpy as np
import matplotlib.pyplot as plt
from pymoo.algorithms.moo.nsga2 import NSGA2
from pymoo.factory import get_problem
from pymoo.optimize import minimize
from pymoo.visualization.scatter import Scatter
```

```python
def run_algorithm(problem_name, iters, N):
    problem = get_problem(problem_name)

    algorithm = NSGA2(pop_size=N)

    algorithm.setup(problem, ('n_gen', iters), seed=1)
    
    plot_frequency = N / 5
    i = 0
    
    plot = Scatter()
    while algorithm.has_next():
        algorithm.next()
        if i % plot_frequency == 0:
            plot.add(algorithm.opt.get("F"), color="green", alpha=0.7)
        
    res=algorithm.result()
    plot.add(problem.pareto_front(), plot_type="line", color="black", alpha=0.7)
    plot.add(res.F, facecolor="none", edgecolor="red")
    plot.show()
    plt.show()
```

```python
run_algorithm("zdt1", 500, 100)
```

```python
run_algorithm("zdt2", 500, 100)
```

```python
run_algorithm("zdt3", 500, 100)
```
